# Estimasi COVID-19 Indonesia Menggunakan Regresi Linear

Project Data Mining - Metode Estimasi (CRISP-DM)

## 1. Import Library

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
```

### Penjelasan Cell Import Library
Cell ini digunakan untuk mengimpor seluruh library yang dibutuhkan dalam project.
- Pandas: mengolah dataset berbentuk tabel.
- NumPy: perhitungan numerik dan statistik.
- Matplotlib: membuat grafik visualisasi.
- Seaborn: visualisasi data yang lebih menarik.
- LinearRegression: algoritma estimasi yang digunakan.
- Metrics: menghitung MAE, RMSE, dan R² Score untuk evaluasi model.

## 2. Load Dataset

```
df = pd.read_excel('ESTIMASI DATA_SEAT(3).xlsx')
print(df.shape)
df.head()
```

### Penjelasan Cell Load Dataset
Cell ini membaca dataset COVID-19 dari file Excel menggunakan `pd.read_excel()`.
`df.shape` digunakan untuk melihat jumlah baris dan kolom dataset.
`df.head()` digunakan untuk menampilkan 5 data pertama sebagai proses pengecekan awal dataset.

## 3. Data Understanding

```
df.info()
df.describe()
```

### Penjelasan Cell Data Understanding
`df.info()` digunakan untuk melihat tipe data, jumlah data, dan missing value.
`df.describe()` digunakan untuk menampilkan statistik deskriptif seperti mean, minimum, maksimum, dan standar deviasi.

## 4. Data Cleaning

```
print(df.isnull().sum())
print('Duplikat:', df.duplicated().sum())
```

### Penjelasan Cell Data Cleaning
`isnull().sum()` digunakan untuk mengecek data kosong.
`duplicated().sum()` digunakan untuk mengecek data yang duplikat.
Tahap ini bertujuan menjaga kualitas data sebelum dilakukan analisis.

## 5. Pemilihan Variabel

```
data = df[['Total Cases','Total Deaths']].dropna()
data.head()
```

### Penjelasan Cell Pemilihan Variabel
Pada penelitian ini hanya digunakan dua atribut yaitu:
- Total Cases sebagai variabel independen (X)
- Total Deaths sebagai variabel dependen (Y)
Fungsi `dropna()` digunakan untuk menghapus baris yang masih memiliki data kosong.

## 6. Korelasi

```
corr = data.corr()
corr
```

### Penjelasan Cell Korelasi
Cell ini menghitung korelasi antar variabel menggunakan `corr()`.
Tujuannya adalah mengetahui kekuatan hubungan antara Total Cases dan Total Deaths sebelum membangun model regresi.

```
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

### Penjelasan Cell Heatmap Korelasi
Heatmap digunakan untuk memvisualisasikan nilai korelasi.
Warna yang semakin merah menunjukkan hubungan positif yang semakin kuat, sedangkan warna biru menunjukkan hubungan negatif.

## 7. Modelling

```
X = data[['Total Cases']]
y = data['Total Deaths']

model = LinearRegression()
model.fit(X,y)

print('Intercept:', model.intercept_)
print('Koefisien:', model.coef_[0])
```

### Penjelasan Cell Modelling
Cell ini membangun model Linear Regression.
- X = Total Cases
- Y = Total Deaths
Model dilatih menggunakan `fit()` sehingga mampu mempelajari hubungan antara jumlah kasus dan jumlah kematian.
Output yang ditampilkan adalah Intercept dan Koefisien regresi.

## 8. Prediksi dan Evaluasi

```
y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print('MAE :', mae)
print('RMSE:', rmse)
print('R2  :', r2)
```

### Penjelasan Cell Evaluasi
`predict()` digunakan untuk menghasilkan nilai prediksi.
MAE mengukur rata-rata kesalahan prediksi.
RMSE mengukur tingkat error model.
R² Score mengukur kemampuan model dalam menjelaskan data. Semakin mendekati 1 maka model semakin baik.

```
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.5)
plt.plot(X, y_pred, linewidth=2)
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Estimasi Total Deaths Berdasarkan Total Cases')
plt.show()
```

### Penjelasan Cell Visualisasi Hasil
Scatter plot menampilkan data aktual.
Garis regresi menampilkan hasil estimasi model.
Semakin dekat titik data terhadap garis regresi maka semakin baik performa model.

## 9. Kesimpulan

- Metode: Regresi Linear
- Target: Total Deaths
- Prediktor: Total Cases
- Evaluasi menggunakan MAE, RMSE, dan R² Score.
